```python
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
y= le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=3)

classifier.fit(x_train,y_train)

y_pred = classifier.predict(x_test)
```

```python
data_p = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```

```python
data_p
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
plt.plot(y_test,y_pred)
plt.ylabel('True label')
plt.xlabel("Predicted label")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score of given model is",accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import precision_score
print("Precision score of given model is",precision_score(y_test,y_pred,average = 'micro')*100)
```

```python
from sklearn.metrics import recall_score
print("Recall score of given model is",recall_score(y_test,y_pred,average = 'micro')*100)
```

```python
from sklearn.metrics import f1_score
print("F1 score of given model is",f1_score(y_test,y_pred,average = 'micro')*100)
```
